```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

concerts = pd.read_excel('/content/GFV_All_Concert_data__10733085.xlsx')
```

```python
concerts.info()
```

```python
#clean date, by getting rid of the day of the week, only including month, day, and year
concerts['Date'] = concerts['Date'].str.slice(start=4)
concerts.head()
```

```python
concerts['Date'] = concerts['Date'].str.replace('\n', ', ')
concerts.head()
```

```python
# clean location, getting rid of weird /n things, and then creating duplicates of city and country
concerts['Location'] = concerts['Location'].str.replace('\n', '')
concerts.head()
```

```python
# make a duplicte column only including country
concerts['Country'] = concerts['Location'].str.split(', ').str[1]
concerts['City'] = concerts['Location'].str.split(',').str[0]
concerts.head()
```
